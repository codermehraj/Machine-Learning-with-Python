# Class Assignment 1: Contor plot of 4k House Data

In the lab we learned the implementation of liniear regression, cost function and gradient descent. And out task was to load the data from 4k house data and make contor plot from the data.

*the following code genarates contorplot from 4k house data*

```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

Reading data from 4k house rent CSV file

```python
data = pd.read_csv('./4K_House_Rent_Dataset.csv')
data.head()
```

```python
# we are selecting house size as the feature
X = data['Size']
Y = data['Rent']

m = len(X)

print("there are total ", data.shape[0], " data records where each records have ", data.shape[1], " features")
```

Now, the data is loaded successfully in array X and Y. Now to generate contor plot we need to define loss functions. which is:

```python
# defining loss function
def loss(a,b):
  return 1/len(X)*(np.sum((a*X + b - Y)**2))
```

Now we can produce the plot. To produce the plot I am using this range:

- X (-400 to 400)
- Y (-300000 to 300000)

```python
# Produce the plots
fig = plt.figure(figsize=(8, 8))

# Contour plot big scope
ax = fig.add_subplot(1, 1, 1, projection="3d")

a = np.arange(-400, 400, 5)
b = np.arange(-300000, 300000, 20000)

A, B = np.meshgrid(a, b)
zs = np.array([loss(a, b) for a, b in zip(np.ravel(A), np.ravel(B))])
loss_vals = zs.reshape(A.shape)

contour = ax.contour(A, B, loss_vals, 30, cmap='cividis', linestyles="solid", offset=0)
ax.plot_surface(A, B, loss_vals, cmap="cividis", rstride=3, cstride=3, edgecolor='none', alpha=0.8)

ax.view_init(40, 290)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('loss value', rotation="vertical")
ax.set_title('Contour Plot Big')
plt.tick_params(axis='both', which='major', labelsize=7)

plt.show()
```

The above plot is not centered properly. To centralize the plot I am using a smaller range:

- X (-100 to 200)
- Y (-180000 to 150000)

```python
fig = plt.figure(figsize=(8, 8))

# Contour plot small scope
ax = fig.add_subplot(1, 1, 1, projection="3d")
a = np.arange(-100, 200, 5)
b = np.arange(-180000, 150000, 15000)
A, B = np.meshgrid(a, b)
zs = np.array([loss(a, b) for a, b in zip(np.ravel(A), np.ravel(B))])
loss_vals = zs.reshape(A.shape)

contour = ax.contour(A, B, loss_vals, 20, cmap='plasma', linestyles="solid", offset=0)
ax.plot_surface(A, B, loss_vals, cmap="plasma", rstride=3, cstride=3, edgecolor='none', alpha=0.8)
ax.view_init(40, 290)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('loss value', rotation="vertical")
ax.set_title('Contour Plot Small')
plt.tick_params(axis='both', which='major', labelsize=7)

plt.show()
```
